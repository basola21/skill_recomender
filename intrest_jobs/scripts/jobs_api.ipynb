```python
import requests
import pandas as pd

def search_jobs():
    app_id = ''  # Replace with your App ID
    app_key = ''  # Replace with your App Key
    results_per_page = 10
    page_number = 1
    total_pages = None
    jobs = []

    while total_pages is None or page_number <= total_pages:
        # Build the API request URL with pagination parameters
        url = f'https://api.adzuna.com/v1/api/jobs/in/search/{page_number}?app_id={app_id}&app_key={app_key}&results_per_page={results_per_page}'

        # Send the API request
        response = requests.get(url)

        # Check if the request was successful (HTTP status code 200)
        if response.status_code == 200:
            # Parse the JSON response
            data = response.json()

            # Extract job information
            if 'results' in data:
                jobs.extend(data['results'])
                total_pages = data['count'] // results_per_page

                if len(jobs)>100:
                    break
                print('-----------------')
                print(f'Page {page_number} of {total_pages}')
                print(f'Found {data["count"]} jobs')
                print(f'Current list size: {len(jobs)}')
                print('-----------------')

                page_number += 1
            else:
                break
        else:
            break
    return jobs

# Usage example
jobs = search_jobs()

# Create a DataFrame with job details
df_jobs = pd.DataFrame(jobs)
```
